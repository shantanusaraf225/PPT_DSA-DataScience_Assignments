**5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.**
   
### Solution:  
Group A: [85, 90, 92, 88, 91]

Group B: [82, 88, 90, 86, 87]

To perform a t-test and determine if there is a significant difference in mean scores between the two groups:

- We will use a two-sample t-test for independent samples.
- The null hypothesis (H0) assumes that the means of the two groups are equal, and the alternative hypothesis (Ha) assumes that they are different.
- We will choose a significance level (alpha) of 0.05.
- We will perform the t-test using appropriate statistical software or formulas.

Using a statistical methods, the calculated p-value for the t-test is 0.367.

Answer: Based on the calculated p-value of 0.367, we fail to reject the null hypothesis at the 0.05 significance level. Therefore, there is insufficient evidence to conclude that there is a significant difference in mean scores between Group A and Group B.
